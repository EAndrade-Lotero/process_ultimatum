## Analysis of times

```python
import re 
import ast
import pandas as pd

from pathlib import Path

from pre_processing_utils import Utils
```

```python
# folder_path = Path(r"/Users/edgarjoseandradelotero/Library/CloudStorage/GoogleDrive-andradeed78@gmail.com/My Drive/UCDavis/Nested games/data/new")
folder_path = Path.cwd().parent / "data" / "new"

dict_worker_id = {}
for subdir in folder_path.iterdir():
    if subdir.is_dir():
        df = pd.read_csv(subdir / "Request.csv")
        worker_id_list = []
        for key, row in df.iterrows():
            value = row['unique_id']
            if isinstance(value, str):
                value_ = value.split(':')[0]
                if value_ not in worker_id_list:
                    worker_id_list.append(value_)
        dict_worker_id[subdir.name] = {
            idx  + 1: value for idx, value in enumerate(worker_id_list)
        }
dict_worker_id
```

```python
# folder_path = Path(r"/Users/edgarjoseandradelotero/Library/CloudStorage/GoogleDrive-andradeed78@gmail.com/My Drive/UCDavis/Nested games/data/new")
folder_path = Path.cwd().parent / "data" / "new"

df_list = []
for subdir in folder_path.iterdir():
    if subdir.is_dir():
        df = pd.read_csv(subdir / "Response.csv")
        df['batch'] = subdir.name
        df_list.append(df)

df = pd.concat(df_list, ignore_index=True)
print(df.shape)
df.head(2)
```

```python
df["time_taken"] = df['metadata_'].apply(
    lambda x: Utils.str_to_dict(x)["time_taken"]
)
df.head(2)
```

```python
# df_times = df.groupby(['batch', 'participant_id', 'page_type'])['time_taken'].describe().reset_index()
df_times = df.groupby(['batch', 'page_type'])['time_taken'].describe()
df_times
```

```python
round_time = 0
for key, time_taken in df.groupby(['batch', 'page_type'])['time_taken']:
    batch, page_type = key
    if page_type in [
            'ConsentPageWrapper', 'GuessingCoinsGamePage',
            'TimedModularPage ', 'InnerWaitingPage '
            'OuterProposalPage	', 'OuterAcceptancePage', 'OuterWaitingPage '
            'InnerProposalPage ', 'InnerAcceptancePage ', 'ScorePage '
        ]:
        round_time += time_taken.mean()
print(f"Average time taken per round = {Utils.format_time(round_time)}")
print(f"Estimated total time based on average = {Utils.format_time(round_time * 25)}")
```

```python
df.groupby(['batch'])['time_taken'].describe()
```

```python
total_time = df.groupby(['batch', 'participant_id'])['time_taken'].sum().mean()
print(f"Average total time = {Utils.format_time(total_time)}")
```

```python
HOURLY_PAYMENT = 25
SECONDS_PAYMENT = HOURLY_PAYMENT / (60 * 60)

for key, grp in df.groupby(['batch', 'participant_id']):
    batch, participant_id = key
    total_time = grp['time_taken'].sum()
    print(f'Id: {participant_id} --- Worker {dict_worker_id[batch][participant_id]} => {Utils.format_time(total_time)} time taken -> payment: {round(SECONDS_PAYMENT * total_time, 2)}')
```
